```python
import numpy as np
import pandas as pd
import seaborn as sns

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize']= (12,8)
```

```python
df = pd.read_csv('movies.csv')
df.head()
```

```python
# checking for the % of missing columns 
for col in df.columns:
    percent_missing = np.mean(df[col].isnull())
    print('{}-{}%'.format(col, percent_missing))
```

```python
df.dtypes
```

```python
# filling NaN values with zero for conversion from float to int
df['budget'].fillna(0, inplace=True)
df['gross'].fillna(0, inplace=True)
```

```python
df['budget']= df['budget'].astype('int64')
df['gross']= df['gross'].astype('int64')
```

```python
df['budget'].dtypes
```

```python
df
```

```python
df.sort_values(by=['gross'], ascending=False)
```

```python
#scatter plot to check for correlation between budget and gross earnings
plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Correlation Between Budget and Gross')

plt.xlabel('Budget for film')

plt.ylabel('Gross Earnings')


plt.show()
```

```python
sns.regplot(x='budget', y='gross', data = df, scatter_kws={'color':'red'}, line_kws={'color': 'blue'});
```

```python
df.corr()
```

```python
correlation_matrix= df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation_matrix for Numeric figures')

plt.xlabel('Movie features')

plt.ylabel('Movie features')

plt.show()
```

```python
# correlation for non numericalvalues
# categorising non_numrical columns
df_num = df

for col_name in df_num.columns:
    if (df_num[col_name].dtype =='object'):
        df_num[col_name]= df_num[col_name].astype('category')
        df_num[col_name]= df_num[col_name].cat.codes
        
df_num
```

```python
correlation_matrix= df_num.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation_matrix for Numeric figures')

plt.xlabel('Movie features')

plt.ylabel('Movie features')

plt.show()
```

```python
correlation_matrix2 = df_num.corr()
corr_pairs = correlation_matrix2.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr= sorted_pairs[(sorted_pairs) > 0.5 ]
high_corr
```
